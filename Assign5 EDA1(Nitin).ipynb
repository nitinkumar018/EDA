# IQR:
Q1(25 percentile) and Q3(75 percentile) => IQR (Inter Quartile Range) = Q3 - Q1. Percentile is calculated using quantile function from pandas library

```python
Q1 = data1.quantile(0.25) #25 percentile of each numerical column is calculated and save in var. Q1 (first 25% of the data)
```

```python
Q1
```

```python
Q3=data1.quantile(0.75) #75 percentile of each numerical column is calculated and save in var. Q3 (first 75% of the data)
```

```python
Q3
```

```python
IQR= Q3-Q1 # IQR of each numerical column is calculated and save in var. IQR
```

```python
IQR
```

Range of middle 50% of the data is given by IQR. For example: let us take Column LB=> first arrange all values of LB in ascending order , then take first 75% of the data and find the maximum value which is 140(calculated above). Now take first 25% of the data and find the maximum value which is 126(calculated above), obviously for middle 50% data 126 will be the minimum value and 140 will be the maximum value. Therefore, range of middle 50% data will be 140-126=14 (also calculated above) which is also known as IQR (Inter Quartile Range) 

# Highlight any interesting findings from this summary

Columns like LB,AC,FM,UC,DL,DS,DP,MSTV,Tendency,NSP have perfectly normal distributed data (because, mean = median = mode)

Column ASTV has left skewed data i.e. majority of data lies in the right side of the mean (because, mode > median > mean)

Columns like ALTV,MLTV,Width have right skewed data i.e. majority of data lies in the left side of the mean 
(because, mean > median > mode)

If std is more => data is more spread about the mean i.e data points are more far away from the mean. Clearly, for Width,ALTV,ASTV  data is relatively more spread about the mean and for DS, DP data is least spread about the mean. 

If IQR is more => middle 50% data is more spread and if IQR is less => middle 50% data cluster more tightly. Clearly, for Width,ASTV,LB,ALTV middle 50% data is relatively more spread and for DS, DP, NSP middle 50% data cluster more tightly

Let us first explore relationship between two variables using Scatter plot and then using heatmap

```python
xpoints=data1['LB']    # let's explore the relation between variables LB and NSP 
ypoints=data1['NSP']   
plt.title('scatterplot')  # title of the graph
plt.xlabel('LB')   # x-axis labelled as LB
plt.ylabel('NSP')   # y-axis labelled as NSP
plt.plot(xpoints,ypoints,'o')  # drawing scatter plot using plot function
plt.show()   #show the graph (it's optional in jupyter notebook)
```

for visualizing the relationship between two variables in this dataset,  we are getting similar pattern of graph using scatter plot. Let's apply heatmap over this dataset to get the exact value of correlation between two variables

```python
fig, ax = plt.subplots(figsize=(10, 10))  #specify size of heatmap
sns.heatmap(data1.corr(),annot=True) #create heatmap (parameter annot is used to display the correlation values, bidefault it is False, so, we have to pass True)
```

DP, ASTV are the most contributing elements which is increasing the NSP i.e. DP, ASTV are the variables which affect the NSP most because correlation value between DP and NSP is 0.47 which is maximum value shown in this heatmap. Likewise, correlation value between ASTV and NSP is also 0.47. Hence, they have positive correlation (both variable varies in same direction)

any value < 0.1 is considered as no-correlation. Therefore, variables like FM, DL have no correlation with NSP. Remaining variables either have negative correlation or very less correlation with NSP (Negative correlation means both variable varies in opposite direction)

# Employ advanced visualization techniques like pair plots, or violin plots for deeper insights.

Violin plot is similar to boxplot. A violin plot shows peaks in the data. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.

A violin plot can be useful when you want to show the shape of the distribution of the data in addition to the information provided by a box plot. It's also useful when you have a larger dataset, as it can show more information about the data distribution.

violin plots show the shape of a data set by using a Probability Density Function (PDF), or a density plot. The width of the PDF describes how frequently that value occurs in the data set. The wider regions of the density plot indicate values that occur more frequently. Narrower regions of the density plot indicate values that occur less frequently.In addition to showing the shape of a data set, a violin plot also summarizes a data set using five summary values: the minimum, first quartile, median, third quartile, and maximum. 

Since, our dataset has only numerical variables. Therefore, we will bring insights of one numerical column at a time using Violin plot

```python
sns.lineplot(x=data1['LB'], y=data1['NSP'],label='pattern', color='blue')
 
# Adding labels and title
plt.xlabel('LB')
plt.ylabel('NSP')
plt.title('checking trend')
 
plt.show()
```


if the slope is a +ve value --> increasing trend

if the slope is a -ve value --> decreasing trend

if the slope is a zero value --> No trend

# 5. Conclusion:

A machine learning algorithm can not run if there are missing values, outliers in the dataset. We can treat missing values using Imputation technique. Imputation technique is used for replacing the missing data with some substitute values to retain most of the data/information of the dataset. These substitute values can be either mean, median or mode. After treating missing values, our next goal is to remove the outliers and for it, outlier detection is needed which is done by boxplot. Once we remove outliers, our data becomes normal distributed i.e. symmetrical about the mean and now, we are in the position to train our machine learning model successfully. Through IQR, we will understand how middle 50% data is spread and heatmap will help us to understand whether our variables have good correlation or not which ultimately helps us in getting close to the future predictions especially when we are working on Time Series data
